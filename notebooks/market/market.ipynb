```python
from asyncio import Lock
from dataclasses import dataclass
from typing import List



@dataclass
class User:
    id: int
    username: str
    password: str
    pass

class SecurityOrder:
    ticker: str
    price: float
    quantity: float
    pass

class Security:
    ticker: str
    min_price: float
    max_price: float
    quoted_decimals: float
    
    def process_orders(self, orders: List[SecurityOrder]) -> bool: ...
    pass

class Market:
    securities: dict[str, Security]
    securities_local_mutex: dict[str, Lock]
    securities_local_queue: dict[str, List[SecurityOrder]]
    
    def __init__(self):
        pass
    
    def do_simulation_step(self):
        for ticker, security in self.securities.items():
            with self.securities_local_mutex:
                queue = self.securities_local_queue[ticker]
                self.securities_local_queue[ticker] = list()
                pass
            
            security.process_orders(queue)
        
        
        pass
    
    def insert_security_order(self, order: SecurityOrder) -> bool:
        ticker = order.ticker
        mutex = self.securities_local_mutex.get(ticker)
        if mutex == None:
            return False
        if not ticker in self.securities_local_queue:
            return False
        with mutex:
            self.securities_local_queue[ticker].append(order)
            pass
        return True
    
    pass



```
